```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome(service=Service('driver/chromedriver.exe'))
driver.get('https://vibe.naver.com/chart/total')
```

```python
xpath = """//*[@id="app"]/div[2]/div/div/a[2]"""
driver.find_element(By.XPATH, xpath).click()
```

```python
soup = BeautifulSoup(driver.page_source, 'lxml')
soup
```

```python
driver.quit()
```

```python
song_tags = soup.select('div.title_badge_wrap span a')
singer_tags = soup.select('td.song div.artist_sub')

print(len(song_tags), len(singer_tags))
```

```python
song_tags[:5]
```

```python
songs = []

for song in song_tags:
    songs.append(song.get_text())

songs
```

```python
singer_tags[:5]
```

```python
singers = []

for div in singer_tags:
    singers.append(div.get('title'))

singers
```

```python
print(len(singers))
print(len(songs))
```

```python
for num, (song, singer) in enumerate(zip(songs, singers)):
    print(str(num) + ': ' + song + '/' + singer)
```

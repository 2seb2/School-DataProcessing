```python
#내가 한 코드
rank = []
main_menu = []
cafe_name = []
link_url = []

list_soup = soup.select('div.sammy')

for li_soup in list_soup:
    rank.append(li_soup.select('.sammyRank')[0].text)
    info_str = li_soup.select('.sammyListing')[0].text
    info_list = info_str.split('\n')
    main_menu.append(info_list[0])
    cafe_name.append(info_list[1])
    if li_soup.find('a')['href'][0] == '/': 
        link_url.append('https://www.chicagomag.com/' + li_soup.find('a')['href'])
    else:
        link_url.append(li_soup.find('a')['href'])

for num, (rank, mainmenu, cafename, linkurl) in enumerate(zip(rank, main_menu, cafe_name, link_url)):
    print(str(rank) + ': ' + mainmenu + ', ' + cafename + ', ' + linkurl)
```

```python
#강사님 코드
rank = []
main_menu = []
cafe_name = []
link_url = []

list_soup = soup.select('div.sammy')

for li_soup in list_soup:
    rank.append(li_soup.select('.sammyRank')[0].text)
    info_str = li_soup.select('.sammyListing')[0].text
    info_list = info_str.split('\n')
    main_menu.append(info_list[0])
    cafe_name.append(info_list[1])
    link_url.append(li_soup.find('a')['href'])

print(len(rank), len(main_menu), len(cafe_name), len(link_url))
```

```python
#강사님답안
from tqdm.notebook import tqdm
import time

price = []
address = []

for n in tqdm(df.index):
    html = requests.get(df['URL'][n], headers={"User-Agent":"Mozilla/5.0"})
    time.sleep(1.5)
    tmp = BeautifulSoup(html.text, 'lxml')
    val_list = tmp.select('p.addy')[0].text.split(',')[0].split()
    price.append(val_list[0][:-1])
    address.append(' '.join(val_list[1:]))
```

```python
import googlemaps
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm
import folium

lat = []
lng = []

API_KEY = ''  
gmaps = googlemaps.Client(key=API_KEY)

for n in tqdm(df.index):
    try:
        tmp_add = df['Address'][n] + ' Chicago'
        if tmp_add == 'multiple locations':
            lat.append(np.nan)
            lng.append(np.nan)
        else:
            tmp_map = gmaps.geocode(tmp_add) 
            if tmp_map:
                tmp_loc = tmp_map[0].get('geometry')
                lat.append(tmp_loc['location']['lat'])
                lng.append(tmp_loc['location']['lng'])
            else:
                lat.append(np.nan)
                lng.append(np.nan)
    except Exception as e:
        lat.append(np.nan)
        lng.append(np.nan)
        print(f"에러: {e}")

df['lat'] = lat
df['lng'] = lng

mean_lat = df['lat'].mean()
mean_lng = df['lng'].mean()

map= folium.Map(location=[mean_lat, mean_lng], zoom_start=10)

for n in df.index:
    if not np.isnan(df.loc[n]['lat']) and not np.isnan(df.loc[n]['lng']):
        folium.Marker(
            location=[df.loc[n]['lat'], df.loc[n]['lng']],
            tooltip="Click me!",
            popup=df.loc[n]['Cafe'],
            icon=folium.Icon(color="green")
        ).add_to(map)

map
```

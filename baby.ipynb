```python
from tqdm.notebook import tqdm

years = range(1880, 2011)
pieces = []
columns = ['name', 'sex', 'births']

for year in tqdm(years):
    path = 'data/babynames/yob%d.txt' % year
    frame = pd.read_csv(path, names=columns)
    frame['year'] = year
    pieces.append(frame)

names = pd.concat(pieces, ignore_index=True)
```

```python
names
```

```python
total_births = names.pivot_table('births', index='year', columns='sex', aggfunc='sum')
total_births.head()
```

```python
total_births.tail()
```

```python
import matplotlib.pyplot as plt

total_births.plot(title='Total births by sex and year', figsize=(13, 7))
plt.show()
```

```python
names.head()
```

```python
def add_group(group):
    group['prop'] = group.births / group.births.sum()
    return group

names = names.groupby(['year', 'sex'], as_index=False).apply(add_group)
```

```python
names
```

```python
names.groupby(['year', 'sex']).prop.sum()
```

```python
def get_top1000(group):
    return group.sort_values(by='births', ascending=False)[:1000]

grouped = names.groupby(['year', 'sex'])
top1000 = grouped.apply(get_top1000)

top1000
```

```python
top1000.reset_index(inplace=True, drop=True)
top1000
```

```python
boys = top1000[top1000.sex == 'M']
girls = top1000[top1000.sex == 'F']
```

```python
total_birth = top1000.pivot_table('births', index='year', columns='name', aggfunc='sum')
```

```python
subset = total_birth[['John', 'Harry', 'Mary', 'Marilyn']]
subset.plot(subplots=True, figsize=(12, 10), grid=False, title='Number of births per year')
plt.show()
```

```python
df = boys[boys.year == 1900]
in1900 = df.sort_values(by='prop', ascending=False).prop.cumsum()
in1900.values.searchsorted(0.5) + 1
```

```python
def get_quantile_count(group, q=0.5):
    group = group.sort_values(by='prop', ascending=False)
    return group.prop.cumsum().values.searchsorted(q) + 1

diversity = top1000.groupby(['year', 'sex']).apply(get_quantile_count)
```

```python
diversity
```

```python
diversity = diversity.unstack('sex')
diversity
```

```python
diversity.plot(title='Number of popular names in top 50%', figsize=(10, 8))
plt.show()
```

```python
get_last_letter = lambda x: x[-1]
last_letters = names.name.map(get_last_letter)
last_letters.name = 'last_letter'
last_letters
```

```python
table = names.pivot_table('births', index=last_letters, columns=['sex', 'year'], aggfunc='sum')
table
```

```python
subtable = table.reindex(columns=[1910, 1960, 2010], level='year')
subtable.head()
```

```python
subtable.sum()
```

```python
letter_prop = subtable / subtable.sum()
letter_prop
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 1, figsize=(10, 8))
plt.subplots_adjust(hspace=0.25)
letter_prop['M'].plot(kind='bar', rot=0, ax=axes[0], title='Male')
letter_prop['F'].plot(kind='bar', rot=0, ax=axes[1], title='Female', legend=False)
plt.show()
```

```python
x = np.arange(0, 10, 0.1)
y1 = 0.3*(x-5)**2 + 1
y2 = -1.5*x + 3
y3 = np.sin(x)**2 # NumPy sin() np.sin() 
y4 = 10*np.exp(-x) + 1 # NumPy exp() np.exp() 

# 2  2     p   
plt.subplot(2,2,1) # p 1
plt.plot(x,y1)

plt.subplot(2,2,2) # p 2
plt.plot(x,y2)

plt.subplot(2,2,3) # p 3
plt.plot(x,y3)

plt.subplot(2,2,4) # p 4
plt.plot(x,y4)

plt.show()
```

```python
city = ['', '', '', '', '', '', '']

# (latitude) (longitude)
lat  = [37.56, 37.45, 36.35, 35.87, 35.53, 35.18, 35.16]
lon = [126.97, 126.70, 127.38, 128.60, 129.31, 129.07, 126.85]

#  (/km^2): 2017  
pop_den = [16154, 2751, 2839, 2790, 1099, 4454, 2995]

size = np.array(pop_den) * 0.2 #   
colors = ['r', 'g', 'b', 'c', 'm', 'k', 'y'] #   

plt.scatter(lon, lat, s=size, c=colors, alpha=0.5)
plt.xlabel('(longitude)')
plt.ylabel('(latitude)')
plt.title('  (2017)')

for x, y, name in zip(lon, lat, city):
    plt.text(x, y, name) #      

plt.show()
```

```python
member_IDs = ['m_01', 'm_02', 'm_03', 'm_04'] #  ID
before_ex = [27, 35, 40, 33] #   
after_ex = [30, 38, 42, 37] #    

n_data = len(member_IDs)     #       4
index = np.arange(n_data)   # NumPy    [0, 1, 2, 3]
plt.bar(index, before_ex)   # bar(x,y) x=index, height=before_ex  
plt.xticks(fontname = "Arial")
plt.yticks(fontname = "Arial")
plt.show()
```

```python
plt.bar(index, before_ex, tick_label=member_IDs)
plt.xticks(fontname = "Arial")
plt.yticks(fontname = "Arial")
plt.show()
```

```python
colors=['tab:red', 'tab:green', 'tab:blue', 'tab:purple']
plt.bar(index, before_ex, tick_label=member_IDs, color=colors)
plt.xticks(fontname = "Arial")
plt.yticks(fontname = "Arial")
plt.show()
```

```python
colors=['tab:red', 'tab:green', 'tab:blue', 'tab:purple']
plt.bar(index, before_ex, tick_label=member_IDs, color=colors, width=0.6)
plt.xticks(fontname = "Arial")
plt.yticks(fontname = "Arial")
plt.show()
```

```python
colors=['tab:red', 'tab:green', 'tab:blue', 'tab:purple']
plt.barh(index, before_ex, tick_label=member_IDs, color=colors)
plt.xticks(fontname = "Arial")
plt.yticks(fontname = "Arial")
plt.show()
```

```python
member_IDs = ['m_01', 'm_02', 'm_03', 'm_04'] #  ID
before_ex = [27, 35, 40, 33] #   
after_ex = [30, 38, 42, 37] #    

n_data = len(member_IDs)     #       4
index = np.arange(n_data)   # NumPy    (0, 1, 2, 3)
colors=['tab:red', 'tab:green', 'tab:blue', 'tab:purple']
bar_width = 0.4
plt.bar(index, before_ex, color='c', align='edge', width = bar_width, label='before')
plt.bar(index + bar_width, after_ex , color='m', align='edge', width = bar_width, label='after')

plt.xticks(index + bar_width, member_IDs)
plt.legend()
plt.xlabel(' ID')
plt.ylabel(' ')
plt.title('    ()  ')
plt.yticks(fontname = "Arial")
plt.xticks(fontname = "Arial")
plt.show()
```

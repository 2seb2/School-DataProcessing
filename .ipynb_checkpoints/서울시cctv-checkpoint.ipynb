```python
import numpy as np
import pandas as pd
```

```python
cctv_seoul = pd.read_csv('./data/서울시 CCTV 설치운영 현황(자치구)-연도별.csv', 
                         header=1, thousands=',', encoding='cp949')
cctv_seoul.head()
```

```python
cctv_seoul.info()
```

```python
cctv_seoul.head()
```

```python
cctv_seoul.drop([0], inplace=True)
cctv_seoul.head()
```

```python
cctv_seoul.rename(columns={cctv_seoul.columns[0]:'구별'}, inplace=True)
cctv_seoul.head()
```

```python
pop_seoul = pd.read_csv('./data/population.txt',
                        header=2, sep='\s+', thousands=',')
pop_seoul.head()
```

```python
pop_seoul.drop(['기간', '세대', '남자', '여자', '남자.1', '여자.1', '남자.2', '여자.2', '세대당인구'],
              axis=1, inplace=True)
pop_seoul.head()
```

```python
pop_seoul.rename(columns={pop_seoul.columns[0]:'구별',
                          pop_seoul.columns[1]:'인구수',
                          pop_seoul.columns[2]:'한국인',
                          pop_seoul.columns[3]:'외국인',
                          pop_seoul.columns[4]:'고령자'}, inplace=True)
pop_seoul.head()
```

```python
pop_seoul.drop([0], inplace=True)
pop_seoul.head()
```

```python
cctv_seoul['구별'].unique()
```

```python
cctv_seoul.loc[cctv_seoul['구별'] == '중 구', '구별'] = '중구'
#cctv_seoul.loc[1, 0] = '중구'
cctv_seoul['구별'].unique()
```

```python
cctv_seoul['최근증가율'] = (cctv_seoul['2016년'] +
                       cctv_seoul['2017년'] +
                       cctv_seoul['2018년'] +
                       cctv_seoul['2019년'] +
                       cctv_seoul['2020년']) / \
                    (cctv_seoul['2011년 이전'] +
                    cctv_seoul['2011년'] +
                    cctv_seoul['2012년'] +
                    cctv_seoul['2013년'] +
                    cctv_seoul['2014년'] + 
                    cctv_seoul['2015년']) * 100

cctv_seoul.sort_values(by='최근증가율', ascending=False)
```

```python
pop_seoul.head()
```

```python
pop_seoul['외국인비율'] = (pop_seoul['외국인'] / pop_seoul['인구수']) * 100
pop_seoul['고령자비율'] = (pop_seoul['고령자'] / pop_seoul['인구수']) * 100
pop_seoul.head()
```

```python
pop_seoul.sort_values(by='인구수', ascending=False)
```

```python
pop_seoul.sort_values(by='고령자', ascending=False)
```

```python
pop_seoul.sort_values(by='고령자비율', ascending=False)
```

```python
data_result = pd.merge(cctv_seoul, pop_seoul, on='구별')
data_result.head()
```

```python
data_result.drop(['2011년 이전', '2011년', '2012년', '2013년', '2014년', '2015년',
                   '2016년', '2017년', '2018년', '2019년', '2020년'],
                 axis=1, inplace=True)
data_result.head()
```

```python
data_result.set_index('구별', inplace=True)
data_result.head()
```

```python
data_result.to_csv('data/cctv_pop_seoul.csv', sep=',', encoding='utf-8')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.pairplot(data_result, x_vars=['고령자비율', '외국인비율', '인구수'],
             y_vars=['총계'], kind='reg', height=3)#height / inch
plt.show()
```

```python
data_result['총계'].plot(kind='barh', grid=True, figsize=(10, 10))
plt.show()
```

```python
#내 답안
data_result = data_result.sort_values(by='총계', ascending=True)
data_result['총계'].plot(kind='barh', grid=True, figsize=(10, 10))
plt.show()
```

```python
#강사님 답안
#ascending=True인 이유는 그래프를 아래부터 그리기 때문에 true로 함.

data_result['총계'].sort_values().plot(kind='barh', grid=True, figsize=(10, 10))
plt.show()
```

```python
data_result['cctv비율'] = data_result['총계'] / data_result['인구수'] * 100
data_result['cctv비율'].sort_values().plot(kind='barh', grid=True, figsize=(10, 10))
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(data_result['인구수'], data_result['총계'], s=50)
plt.xlabel('인구수')
plt.ylabel('cctv')
plt.show()
```

```python
fp = np.polyfit(data_result['인구수'], data_result['총계'], 1)
fp
```

```python
f1 = np.poly1d(fp)
fx = np.linspace(100000, 700000, 100)
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(data_result['인구수'], data_result['총계'], s=50)
plt.plot(fx, f1(fx), ls='dashed', color='g')
plt.xlabel('인구수')
plt.ylabel('cctv')
plt.grid()
plt.show()
```

```python
data_result['오차'] = np.abs(data_result['총계'] - f1(data_result['인구수']))
df_sort = data_result.sort_values(by='오차', ascending=False)
df_sort.head()
```

```python
plt.figure(figsize=(14,10))
plt.scatter(data_result['인구수'], data_result['총계'], c=data_result['오차'], s=50)
plt.plot(fx, f1(fx), ls='dashed', color='g')

for n in range(10):
    plt.text(df_sort['인구수'].iloc[n]*1.02, df_sort['총계'].iloc[n]*0.98, 
             df_sort.index[n], fontsize=15)
    
plt.xlabel('인구수')
plt.ylabel('cctv')
plt.colorbar()
plt.grid()
plt.show()
```

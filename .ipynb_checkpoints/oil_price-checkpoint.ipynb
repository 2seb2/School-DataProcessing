```python
driver = webdriver.Chrome(service=Service('driver/chromedriver.exe'))
driver.get('https://www.opinet.co.kr/searRgSelect.do')
```

```python
gu_list = gu_list_raw.find_elements(By.TAG_NAME, 'option')
```

```python
gu_names = [ option.get_attribute('value') for option in gu_list ]
gu_names
```

```python
print(len(gu_names))
```

```python
gu_names = gu_names[1:]
gu_names
```

```python
element = driver.find_element(By.XPATH, xpath)
element.send_keys(gu_names[0])
```

```python
import time
from tqdm.notebook import tqdm
```

```python
import pandas as pd
from glob import glob
```

```python
glob('data/oil_price/*.xls')
```

```python
stations_files = glob('data/oil_price/*.xls')

tmp_raw = []
for file_name in stations_files:
    tmp = pd.read_excel(file_name, header=2)
    tmp_raw.append(tmp)
```

```python
print(len(tmp_raw))
```

```python
station_raw = pd.concat(tmp_raw)
station_raw.info()
```

```python
station_raw.head(10)
```

```python
stations = pd.DataFrame({'Oil_store': station_raw['상호']
                        , '주소': station_raw['주소']
                        , '가격': station_raw['휘발유']
                        , '셀프': station_raw['셀프여부']
                        , '상표': station_raw['상표']})
stations.head(10)
```

```python
stations['구'] = [ address.split()[1] for address in stations['주소'] ]
stations.head(10)
```

```python
stations['구'].unique
```

```python
stations[stations['가격'] == '-']
```

```python
oil_price_top10 = stations.sort_values(by='가격', ascending=False).head(10)
oil_price_top10
```

```python
oil_price_bottom10 = stations.sort_values(by='가격', ascending=True).head(10)
oil_price_bottom10
```

```python
import googlemaps
gmaps_key = ''
gmaps = googlemaps.Client(key=gmaps_key)
```

```python
import numpy as np
```

```python
from tqdm.notebook import tqdm

lat = []
lng = []

for n in tqdm(oil_price_top10.index):
    try:
        tmp_add = oil_price_top10['주소'][n].split('(')[0]
        tmp_map = gmaps.geocode(tmp_add)
        tmp_loc = tmp_map[0].get('geometry')
        lat.append(tmp_loc['location']['lat'])
        lng.append(tmp_loc['location']['lng'])
    except:
        lat.append(np.nan)
        lng.append(np.nan)
        print('nan!!')
```

```python
oil_price_top10['lat'] = lat
oil_price_top10['lng'] = lng
oil_price_top10
```

```python
from tqdm.notebook import tqdm

lat = []
lng = []

for n in tqdm(oil_price_bottom10.index):
    try:
        tmp_add = oil_price_bottom10['주소'][n].split('(')[0]
        tmp_map = gmaps.geocode(tmp_add)
        tmp_loc = tmp_map[0].get('geometry')
        lat.append(tmp_loc['location']['lat'])
        lng.append(tmp_loc['location']['lng'])
    except:
        lat.append(np.nan)
        lng.append(np.nan)
        print('nan!!')
```

```python
oil_price_bottom10['lat'] = lat
oil_price_bottom10['lng'] = lng
oil_price_bottom10
```

```python
map = folium.Map(location=[37.5502, 126.982], zoom_start=11)

for n in oil_price_top10.index:
    folium.CircleMarker([oil_price_top10['lat'][n], oil_price_top10['lng'][n]]
                 , radius=15
                 , color='#cd3181'
                 , fill_color='#cd3181'
                 , fill=True).add_to(map)

for n in oil_price_bottom10.index:
    folium.CircleMarker([oil_price_bottom10['lat'][n], oil_price_bottom10['lng'][n]]
                 , radius=15
                 , color='#3186cc'
                 , fill_color='#3186cc'
                 , fill=True).add_to(map)

map
```

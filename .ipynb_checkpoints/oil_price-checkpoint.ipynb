```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome(service=Service('driver/chromedriver.exe'))
driver.get('https://www.opinet.co.kr/searRgSelect.do')
```

```python
xpath = """//*[@id="SIGUNGU_NM0"]"""
gu_list_raw = driver.find_element(By.XPATH, xpath)
```

```python
print(gu_list_raw)
```

```python
gu_list = gu_list_raw.find_elements(By.TAG_NAME, 'option')
```

```python
gu_names = [ option.get_attribute('value') for option in gu_list ]
gu_names
```

```python
print(len(gu_names))
```

```python
gu_names = gu_names[1:]
gu_names
```

```python
element = driver.find_element(By.XPATH, xpath)
element.send_keys(gu_names[0])
```

```python
btn_xpath = """//*[@id="templ_list0"]/div[7]/div/a"""
driver.find_element(By.XPATH, btn_xpath).click()
```

```python
import time
from tqdm.notebook import tqdm
```

```python
for gu in tqdm(gu_names):
    xpath = """//*[@id="SIGUNGU_NM0"]"""
    element = driver.find_element(By.XPATH, xpath)
    element.send_keys(gu)
    time.sleep(3)
    btn_xpath = """//*[@id="templ_list0"]/div[7]/div/a"""
    driver.find_element(By.XPATH, btn_xpath).click()
    time.sleep(3)
```

```python
import pandas as pd
from glob import glob
```

```python
glob('data/oil_price/*.xls')
```

```python
stations_files = glob('data/oil_price/*.xls')

tmp_raw = []
for file_name in stations_files:
    tmp = pd.read_excel(file_name, header=2)
    tmp_raw.append(tmp)
```

```python
print(len(tmp_raw))
```

```python
station_raw = pd.concat(tmp_raw)
station_raw.info()
```

```python
station_raw.head(10)
```

```python
stations = pd.DataFrame({'Oil_store': station_raw['상호']
                        , '주소': station_raw['주소']
                        , '가격': station_raw['휘발유']
                        , '셀프': station_raw['셀프여부']
                        , '상표': station_raw['상표']})
stations.head(10)
```

```python
stations['구'] = [ address.split()[1] for address in stations['주소'] ]
stations.head(10)
```
